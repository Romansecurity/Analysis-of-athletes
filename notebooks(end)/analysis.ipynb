```python
import pandas as pd 
```

```python
df = pd.read_excel("/Users/roman/Documents/pythonprogect1/Developer/Analitika/data/forms.xlsx")
df_survey = df.copy()
df_survey.head()
```

```python
df_survey.info()
df_survey.describe()
df_survey.isnull().sum()
```

```python
#df_survey = df_survey.dropna() # удаляем строки с пропущенными значениями
```

```python
print(f"Возраст : {df_survey['Возраст'].sort_values().unique()}")
print(f"Весовая категория: {df_survey['Весовая категория (формат: 62 - 68)'].unique()}")
print(df_survey['Стиль боя:'].unique())
print(df_survey['Любимая техника:'].unique())
```

```python
df_survey['Комфортная дистанция:'] = df_survey['Комфортная дистанция:'].map({
    'Близкая': 1,
    'Средняя': 2,
    'Дальняя': 3
})
df_survey.head(18)
```

```python
df_survey['Как оцениваешь свою защиту?'] = df_survey['Как оцениваешь свою защиту?'].map({
    'Слабая': 1,
    'Средняя': 2,
    'Сильная': 3
})
df_survey.head(18)
```

```python
df_survey['Как у тебя с выносливостью?'] = df_survey['Как у тебя с выносливостью?'].map({
    'Быстро устаю': 1,
    'Комфортно себя чувствую': 2,
    'Очень выносливый/выносливая': 3
})
df_survey.head(18)
```

```python
df_survey['Контроль боя:'] = df_survey['Контроль боя:'].map({
    'Зависит от соперника': 1,
    'Подстраиваешься под противника': 2,
    'Контролируешь бой': 3
})
df_survey.head(18)
```

```python
df_survey['В бою ты чаще...'] = df_survey['В бою ты чаще...'].map({
    'Нервничаешь': 1,
    'Зависит от ситуации': 2,
    'Уверен/Уверена': 3
})
df_survey.head(18)
```

```python
df_survey['Весовая категория (формат: 62 - 68)'] = df_survey['Весовая категория (формат: 62 - 68)'].map({
    '-45': 1,
    '50-55': 2,
    '50-56':2,
    '55-60':3,
    '56-62': 3,
    '57-63':3,
    '62-68':4,
    '63-70': 4,
    '68-75':5,
    '75-':6
})
df_survey.head(18)
```

```python
df_survey.to_excel('/Users/roman/Documents/pythonprogect1/Developer/Analitika/data/survey_table.xlsx', index=False)
```

```python
df_survey.corr(numeric_only=True)

```

```python
df_survey.groupby("Стиль боя:")['Комфортная дистанция:'].mean()

df_survey.groupby("Стиль боя:")['Как оцениваешь свою защиту?'].mean()
df_survey.groupby("Стиль боя:")['Как у тебя с выносливостью?'].mean()
df_survey.groupby("Стиль боя:")['В бою ты чаще...'].mean()
df_survey.groupby("Пол")['В бою ты чаще...'].mean()
df_survey.groupby("Пол")['Как у тебя с выносливостью?'].mean()
df_survey.groupby("Возраст")['Как у тебя с выносливостью?'].mean()
df_survey.groupby("Возраст")['В бою ты чаще...'].mean()

df_survey["Комфортная дистанция:"].corr(df_survey["Контроль боя:"], method='spearman')
df_survey["Весовая категория (формат: 62 - 68)"].corr(df_survey["Как у тебя с выносливостью?"], method='spearman')



```
