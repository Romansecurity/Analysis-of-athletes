```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_excel("/Users/roman/Documents/pythonprogect1/Developer/Analitika/data/forms.xlsx")
df_survey = df.copy()
df_survey.head()
```

```python
df_survey.info()
df_survey.describe()
df_survey.isnull().sum()
```

```python
df_survey = df_survey.dropna() # удаляем строки с пропущенными значениями
```

```python
print(f"Возраст : {df_survey['Возраст'].sort_values().unique()}")
print(f"Весовая категория: {df_survey['Весовая категория (формат: 62 - 68)'].unique()}")
print(df_survey['Стиль боя:'].unique())
print(df_survey['Любимая техника:'].unique())
```

```python
df_survey['Стиль боя:'] = df_survey['Стиль боя:'].map({
    'Универсальный': 1,
    'Атакующий': 2,
    'Контратакующий': 3
})

```

```python
df_survey['Что используешь чаще?'] = df_survey['Что используешь чаще?'].map({
    'Руки': 1,
    'Ноги': 2,
    'Одинаково': 3
})
df_survey.head(18)
```

```python
df_survey['Комфортная дистанция:'] = df_survey['Комфортная дистанция:'].map({
    'Близкая': 1,
    'Средняя': 2,
    'Дальняя': 3
})
df_survey.head(18)
```

```python
df_survey['Как оцениваешь свою защиту?'] = df_survey['Как оцениваешь свою защиту?'].map({
    'Слабая': 1,
    'Средняя': 2,
    'Сильная': 3
})
df_survey.head(18)
```

```python
df_survey['Как у тебя с выносливостью?'] = df_survey['Как у тебя с выносливостью?'].map({
    'Быстро устаю': 1,
    'Комфортно себя чувствую': 2,
    'Очень выносливый/выносливая': 3
})
df_survey.head(18)
```

```python
df_survey['Контроль боя:'] = df_survey['Контроль боя:'].map({
    'Зависит от соперника': 1,
    'Подстраиваешься под противника': 2,
    'Контролируешь бой': 3
})
df_survey.head(18)
```

```python
df_survey['В бою ты чаще...'] = df_survey['В бою ты чаще...'].map({
    'Нервничаешь': 1,
    'Зависит от ситуации': 2,
    'Уверен/Уверена': 3
})
df_survey.head(18)
```
